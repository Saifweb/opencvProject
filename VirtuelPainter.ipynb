```python
import cv2
import time
import numpy as np
#we use os to import folder into our script !
import os 
import HandTrackingModule as htm
```

```python
#import our images!
folderPath="Header"
myList=os.listdir(folderPath)
print(myList)
```

```python
#save image in a list ! 
overLayList=[]
drawColor=(255,0,255)
for imPath in myList:
    #read each image and save it in the list !
    image=cv2.imread(f'{folderPath}/{imPath}')
    overLayList.append(image)
print(len(overLayList))
#header is our img in the screen we will make a default value with image 1 !
header=overLayList[0]

```

```python
#run our com 
cap=cv2.VideoCapture(0)
cap.set(3,1280)
cap.set(4,720)
```

```python
detector=htm.handDetector(detectionCon=0.85) 
imgCanvas = np.zeros((720, 1280, 3), np.uint8)
```

```python
#variable:
xp,yp=0,0
while True:
    # 1.read image 
    succes,img=cap.read()
    #flip the image 
    img=cv2.flip(img,1)
    # 2 find hand landmark
    img=detector.findHands(img)
    lmList,box=detector.findPosition(img,draw=False)
    if (len(lmList)!=0):
        #tip finger
        x1,y1=lmList[8][1:]
        #middle finger
        x2,y2=lmList[11][1:]
        # 3 check which fingers are up 
        fingers=detector.fingersUp()
        #print(fingers)
        # 4--------------------------- selection mode change image---------------------------------------------
        #the condition if finger one and finger 2 are up then do !
        if fingers[1] and fingers[2]:
            xp,yp=0,0
            print("Selection mode")
            cv2.rectangle(img,(x1,y1-25),(x2,y2+25),drawColor,cv2.FILLED)
            #check if we are in the header ! 
            if y1<125:
                #it mean we are clicking first one !
                if 250<x1<450:
                    header=overLayList[0]
                    drawColor=(255,0,255)
                #it mean we are clicking Seconde one !
                elif  550<x1<750:
                    header=overLayList[1]
                    drawColor=(255,0,0)
                #it mean we are clicking third one !
                elif  800<x1<950:
                    header=overLayList[2]
                    drawColor=(0,255,0)
                #it mean we are clicking four one !
                elif  1050<x1<1200:
                    header=overLayList[3]
                    drawColor=(0,0,0)
                
        if all (x >= 1 for x in fingers):
                    xp,yp=0,0
                    cv2.circle(img,(x1,y1),15,(255,255,255),cv2.FILLED)
                    imgCanvas = np.zeros((720, 1280, 3), np.uint8)
        # 5------------------------------------------- drawing mode if finger is up ---------------------------------------------
        if fingers[1] and fingers[2]==False:
            print("Drawing mode")
            cv2.circle(img,(x1,y1),15,drawColor,cv2.FILLED)
            if (xp==0 and yp==0):
                xp,yp=x1,y1
            if drawColor==(0,0,0):
                cv2.line(img,(xp,yp),(x1,y1),drawColor,50)
                cv2.line(imgCanvas,(xp,yp),(x1,y1),drawColor,50)
            else:
                cv2.line(img,(xp,yp),(x1,y1),drawColor,15)
                cv2.line(imgCanvas,(xp,yp),(x1,y1),drawColor,15)
            
            xp,yp=x1,y1
        
    #set header in image
    img[0:125,0:1280]=header
    #set the imgcanvas to our img
    img=cv2.addWeighted(img,0.5,imgCanvas,0.5,0)
    #showing image
    cv2.imshow("image",img)
    cv2.waitKey(1)
```
